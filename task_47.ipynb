````python
import matplotlib.pyplot as plt

A = (1, 3)
B = (5, 4)

plt.figure(figsize=(5,5))

# Plot points
plt.scatter(*A, color='blue')
plt.scatter(*B, color='red')

# Labels
plt.text(A[0], A[1]+0.1, 'A(1,3)')
plt.text(B[0], B[1]+0.1, 'B(5,4)')

# Draw vector AB
plt.quiver(A[0], A[1],
           B[0]-A[0], B[1]-A[1],
           angles='xy', scale_units='xy', scale=1)

plt.xlim(0,6)
plt.ylim(0,5)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Vector AB")

plt.show()
````

````python
import numpy as np

points = np.array([[1, 8, 5],
[8, 8, 2],
[7, 9, 1],
[5, 6, 3],
[9, 1, 2],
[6, 5, 9],
[5, 8, 6],
[3, 4, 8],
[6, 8, 4],
[7, 2, 7]])

query = np.array([3,9,3])

dist = np.linalg.norm(points - query, axis=1)

index = np.argmin(dist)

print("Nearest Point:", points[index])
print("Distance:", dist[index])
````
