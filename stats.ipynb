```python
df = pd.DataFrame()
x = pd.Series([12,25,68,42,113])
y = pd.Series([11,29,58,121,100])
df['x'] = x
df['y'] = y
```

```python
np.cov(df['x'],df['y'])
```

```python
fig,ax=plt.subplots(nrows=1,ncols=3,figsize=(10,5))
ax[0].scatter(df['x'],df['y'])
ax[0].set_title('cov is '+ str(np.cov(df['x'],df['y'])[0][1]))
ax[1].scatter(df['x'],df['x'])
ax[1].set_title('cov is '+ str(np.cov(df['x'],df['y'])[0][0]))
ax[2].scatter(df['y'],df['y'])
ax[2].set_title('cov is '+ str(np.cov(df['x'],df['y'])[1][1]))
```

```python
fig,ax=plt.subplots(nrows=1,ncols=3,figsize=(10,5))
ax[0].scatter(df['x'],df['y'])
ax[0].set_title('corr is '+ str(np.corrcoef(df['x'],df['y'])[0][1]))
ax[1].scatter(df['x'],df['x'])
ax[1].set_title('corr is '+ str(np.corrcoef(df['x'],df['y'])[0][0]))
ax[2].scatter(df['y'],df['y'])
ax[2].set_title('corr is '+ str(np.corrcoef(df['x'],df['y'])[1][1]))
```

```python
import random
l=[]
for i in range(10000):
  x=random.randint(1,6)
  # y=random.randint(1,6)
  l.append(x)
df=pd.Series(l)
s=((df.value_counts())/(df.value_counts().sum())).sort_index(ascending=True)
np.cumsum(s).plot(kind='bar')
```

```python
import random
l=[]
for i in range(10000):
  x=random.randint(1,6)
  y=random.randint(1,6)
  l.append(x+y)
df=pd.Series(l)
s=((df.value_counts())/(df.value_counts().sum())).sort_index(ascending=True)
np.cumsum(s).plot(kind='bar')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import normal
from scipy.stats import norm
sample = normal(loc=50, scale=5, size=1000)

sample_mean = sample.mean()
sample_std = sample.std()

dist = norm(sample_mean, sample_std)

values = np.linspace(sample.min(), sample.max(), 100)

probabilities = dist.pdf(values)

plt.hist(sample, bins=10, density=True)
plt.plot(values, probabilities)
plt.show()
```
