```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import normal
from scipy.stats import norm
sample = normal(loc=50, scale=5, size=1000)

sample_mean = sample.mean()
sample_std = sample.std()

dist = norm(sample_mean, sample_std)

values = np.linspace(sample.min(), sample.max(), 100)

probabilities = dist.pdf(values)

plt.hist(sample, bins=10, density=True)
plt.plot(values, probabilities)
plt.show()
```

```python
df=sns.load_dataset('titanic')
sp=[]
for i in range(100):
  sp.append(df['age'].fillna(df['age'].mean()).sample(30))
sp=np.array(sp)
mean=sp.mean(axis=1)
stnd=sp.std(axis=1).std()
# pd.Series(mean).plot(kind='kde')
mean1=mean.mean()
upper=mean1+(stnd/np.sqrt(30))*2.04
lower=mean1-(stnd/np.sqrt(30))*2.04
print(upper)
print(lower)
print(df['age'].mean())
```

```python
df=sns.load_dataset('titanic')
from scipy.stats import shapiro
# H0 FOR AGE IS MEAN IS 35
# H1 FOR AGE IS MEAN LESS THAN 35
# SAMPLE STD DEVN IS GIVEN alpha is 0.05
df1=df['age'].dropna()
sample_age=df1.sample(25).tolist()
# check for normality using Shapiro Wilk test
from scipy.stats import shapiro

shapiro_age = shapiro(sample_age)

print(shapiro_age)
import scipy.stats as stats

t_statistic, p_value = stats.ttest_1samp(sample_age, 25)

print("t-statistic:", t_statistic)
print("p-value:", p_value/2)
if (p_value/2) < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
