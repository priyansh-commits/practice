```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
df.pivot_table(index='survived',columns='pclass',values='alive',aggfunc='count').plot(kind='bar')

#
```

```python
df = pd.DataFrame()
x = pd.Series([12,25,68,42,113])
y = pd.Series([11,29,58,121,100])
df['x'] = x
df['y'] = y
```

```python
np.cov(df['x'],df['y'])
```

```python
fig,ax=plt.subplots(nrows=1,ncols=3,figsize=(10,5))
ax[0].scatter(df['x'],df['y'])
ax[0].set_title('cov is '+ str(np.cov(df['x'],df['y'])[0][1]))
ax[1].scatter(df['x'],df['x'])
ax[1].set_title('cov is '+ str(np.cov(df['x'],df['y'])[0][0]))
ax[2].scatter(df['y'],df['y'])
ax[2].set_title('cov is '+ str(np.cov(df['x'],df['y'])[1][1]))
```

```python
fig,ax=plt.subplots(nrows=1,ncols=3,figsize=(10,5))
ax[0].scatter(df['x'],df['y'])
ax[0].set_title('corr is '+ str(np.corrcoef(df['x'],df['y'])[0][1]))
ax[1].scatter(df['x'],df['x'])
ax[1].set_title('corr is '+ str(np.corrcoef(df['x'],df['y'])[0][0]))
ax[2].scatter(df['y'],df['y'])
ax[2].set_title('corr is '+ str(np.corrcoef(df['x'],df['y'])[1][1]))
```
