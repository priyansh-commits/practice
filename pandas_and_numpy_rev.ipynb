```python
import numpy as np
import pandas as pd
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)
points_df
def euc(row):
  return (row['1st point'][0]-row['2nd point'][0])**2+(row['1st point'][1]-row['2nd point'][1])**2
points_df['distance']=points_df.apply(euc,axis=1)
points_df
```

```python
import numpy as np
import pandas as pd
course=pd.read_csv('/content/courses.csv')
stud=pd.read_csv('/content/students.csv')
nov=pd.read_csv('/content/reg-month1.csv')
dec=pd.read_csv('/content/reg-month2.csv')
mt=pd.read_csv('/content/matches.csv')
deliveries=pd.read_csv('/content/deliveries.csv')
```

```python
reg=pd.concat([nov,dec],keys=['novemver','decmber'])
reg
```

```python
stud.merge(reg,how='inner',on='student_id')
```

```python
# 1. find total revenue generated
```

```python
course.merge(reg,how='inner',on='course_id')['price'].sum()
```

```python
# 2. find month by month revenue
```

```python
reg1=pd.concat([nov,dec],keys=['novemver','decmber']).reset_index()
course.merge(reg1,how='inner',on='course_id').groupby('level_0')['price'].sum()
```

```python
# 3. Print the registration table
# cols -> name -> course -> price
```

```python
course.merge(reg,how='inner',on='course_id').merge(stud,how='inner',on='student_id')[['name','course_name','price']]
```

```python
# 4. Plot bar chart for revenue/course
course.merge(reg,how='inner',on='course_id').groupby('course_name')['price'].sum().plot(kind='bar')
```

```python
# 5. find students who enrolled in both the months
stud.merge(nov,how='inner',on='student_id').merge(dec,how='inner',on='student_id')['name'].unique()
```

```python
# 6. find course that got no enrollment
# courses['course_id']
# regs['course_id']
op=np.setdiff1d(course['course_id'].unique(),reg['course_id'].unique())

course[course['course_id'].isin(op)]
```

```python
# 7. find students who did not enroll into any courses
stud['student_id'].unique()
reg['student_id'].unique()
ol=np.setdiff1d(stud['student_id'].unique(),reg['student_id'].unique())
stud[stud['student_id'].isin(ol)]
```

```python
# 8. Print student name -> partner name for all enrolled students
# self join
stud.merge(stud,how='inner',left_on='student_id',right_on='partner')[['name_x','name_y']]
```

```python
# 9. find top 3 students who did most number enrollments
stud.merge(reg,how='inner',on='student_id').groupby(['name','student_id'])['name'].count().sort_values(ascending=False).head(3)
```

```python
# 10. find top 3 students who spent most amount of money on courses
stud.merge(reg,how='inner',on='student_id').merge(course,how='inner',on='course_id').groupby(['name','student_id'])['price'].sum().sort_values(ascending=False).head(3)
```
